```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Load the scene (background) and the sample (foreground) images
scene = cv2.imread("img61.jpg")            # Scene image
sample = cv2.imread("Sample_Image.jpg")    # Contains multiple animals/objects on white background

# Check if images loaded correctly
if scene is None or sample is None:
    print("Error: Could not load one or both images. Check file paths!")
    exit()

# 2. Define where in the scene you want to place the sample image
x, y = 50, 200  # Adjust as needed (top-left corner of placement)

# Ensure the sample will fit in the scene at (x, y)
if y + sample.shape[0] > scene.shape[0] or x + sample.shape[1] > scene.shape[1]:
    print("Warning: Sample image does not fit at the specified position. Adjusting to fit inside the scene.")
    # Clamp the sample to fit (optional logic)
    x = min(x, scene.shape[1] - sample.shape[1])
    y = min(y, scene.shape[0] - sample.shape[0])

# 3. Create a mask for the near-white background in the sample image
#    We'll treat pixels close to pure white (e.g., >=240 in each channel) as background.
lower_white = (230, 230, 230)
upper_white = (255, 255, 255)
white_mask = cv2.inRange(sample, lower_white, upper_white)  # single-channel mask (0 or 255)

# Invert mask so objects are 1 where we keep them
objects_mask = cv2.bitwise_not(white_mask)  # 255 where the objects are, 0 where background is

# Convert masks to float32 in [0, 1] for blending
alpha_obj = objects_mask.astype(np.float32) / 255.0      # 1 where we keep the sample image
alpha_bg = 1.0 - alpha_obj                               # 1 where background remains from scene

# Expand these to 3 channels for color arithmetic
alpha_obj_3c = cv2.merge([alpha_obj, alpha_obj, alpha_obj])
alpha_bg_3c = cv2.merge([alpha_bg, alpha_bg, alpha_bg])

# 4. Extract the region of interest (ROI) from the scene where we will place the sample
roi = scene[y : y + sample.shape[0], x : x + sample.shape[1]]

# Convert everything to float32 for blending
roi_f     = roi.astype(np.float32)
sample_f  = sample.astype(np.float32)

# 5. Blend the ROI and the sample image:
#    blended = roi * alpha_bg + sample * alpha_obj
roi_bg    = cv2.multiply(roi_f, alpha_bg_3c)
sample_fg = cv2.multiply(sample_f, alpha_obj_3c)
blended   = cv2.add(roi_bg, sample_fg)

# 6. Put the blended result back into the scene (convert to uint8)
scene[y : y + sample.shape[0], x : x + sample.shape[1]] = blended.astype(np.uint8)

# 7. Show and/or save the composite image
plt.figure(figsize=(10, 6))
plt.imshow(cv2.cvtColor(scene, cv2.COLOR_BGR2RGB))
plt.title("Composite Image")
plt.axis("off")
plt.show()

cv2.imwrite("Composite_Output.jpg", scene)
```

```python
import cv2
import numpy as np

# Load images
scene = cv2.imread("img61.jpg")
sample = cv2.imread("Sample_Image.jpg")

# Resize sample if needed
sample = cv2.resize(sample, (300, 300))  # Example size

# 1. Create mask using broader thresholds for white
lower_white = (200, 200, 200)
upper_white = (255, 255, 255)
mask = cv2.inRange(sample, lower_white, upper_white)

# 2. Morphological operations to clean edges
kernel = np.ones((3,3), np.uint8)
mask = cv2.erode(mask, kernel, iterations=1)
mask = cv2.dilate(mask, kernel, iterations=1)

# 3. Optional: Feather the mask edges by blurring
mask_float = mask.astype(np.float32) / 255.0
mask_float = cv2.GaussianBlur(mask_float, (3, 3), 0)
mask = (mask_float * 255).astype(np.uint8)

# 4. Invert mask and create alpha channels
mask_inv = cv2.bitwise_not(mask)

# 5. Position in the scene (top-left corner at (x, y))
x, y = 50, 200
h_obj, w_obj = sample.shape[:2]
roi = scene[y:y+h_obj, x:x+w_obj]

# Convert everything to float for blending
roi_f = roi.astype(np.float32)
sample_f = sample.astype(np.float32)
mask_f = mask.astype(np.float32) / 255.0
mask_inv_f = mask_inv.astype(np.float32) / 255.0

# Expand to 3 channels
mask3 = cv2.merge([mask_f, mask_f, mask_f])
mask_inv3 = cv2.merge([mask_inv_f, mask_inv_f, mask_inv_f])

# Blend
roi_bg = roi_f * mask_inv3
sample_fg = sample_f * mask3
blended = roi_bg + sample_fg

# Place blended region back
scene[y:y+h_obj, x:x+w_obj] = blended.astype(np.uint8)

cv2.imwrite("Composite_Output.jpg", scene)
```
