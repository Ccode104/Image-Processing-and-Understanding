```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the uploaded Lenna image
image_path = "Lenna.png"
image = cv2.imread(image_path, cv2.IMREAD_COLOR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB

# Define the scaling factor
scale_factor = 8  # Scaling up by 2x

# Apply Nearest Neighbor Interpolation
nearest_neighbor = cv2.resize(image, (image.shape[1] * scale_factor, image.shape[0] * scale_factor), interpolation=cv2.INTER_NEAREST)

# Apply Bilinear Interpolation
bilinear = cv2.resize(image, (image.shape[1] * scale_factor, image.shape[0] * scale_factor), interpolation=cv2.INTER_LINEAR)

# Plot the results
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(image)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(nearest_neighbor)
axes[1].set_title("Nearest Neighbor Interpolation")
axes[1].axis("off")

axes[2].imshow(bilinear)
axes[2].set_title("Bilinear Interpolation")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```

Comparison of Nearest Neighbor vs. Bilinear Interpolation on Lenna Image
Nearest Neighbor Interpolation:

Sharpness: Retains sharp edges but produces blocky artifacts.
Smoothness: Poor; visible jagged edges.
Artifacts: Pixelation and staircase effects on curves and diagonal lines.
Bilinear Interpolation:

Sharpness: Slightly reduced compared to nearest neighbor due to blending.
Smoothness: Improved; color transitions are gradual.
Artifacts: Minimal, but slight blurring occurs.
Conclusion:

Nearest Neighbor is best for preserving individual pixel details (useful in pixel art, low-resolution graphics).
Bilinear Interpolation provides a smoother and more natural appearance, making it preferable for general image scaling.
